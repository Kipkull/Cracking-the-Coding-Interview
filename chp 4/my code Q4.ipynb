```python
def binarytree_to_linkedlist(curr, depth = 1, listdict = {}):
    if curr.content == None:
        return listdict

    cuurNode = listNode(curr.content)

    if str(depth) not in listdict:
        listdict[str(depth)] = linkedlist()
    listdict[str(depth)].put(cuurNode) 
    
    if curr.left != None:
        listdict = binarytree_to_linkedlist(curr.left, depth+1, listdict)
        
    if curr.right != None:
        listdict = binarytree_to_linkedlist(curr.right, depth+1, listdict)
        
    return listdict
```

```python
class binaryTree:
    def __init__(self, content, left=None, right=None):
        self.content = content
        self.left = left
        self.right = right
        
    def __str__(self):
        if self.content == None:
            return 'None'
        return '[' + str(self.content) + \
    ':[' + 'leftChild' + ':' + str(self.left) + ',' \
    'rightChild' + ':' + str(self.right) + ']]'
```

```python
class listNode:
    def __init__(self, value):
        self.value = value
        self.next = None

    def __str__(self):
        return str(self.value)
    
class linkedlist:
    def __init__(self):
        self.head = None
        self.tail = None
    
    def put(self, anote):
        if self.head == None:
            self.head = anote
            self.tail = anote
        else:
            self.tail.next = anote
            self.tail = anote
    
    def __str__(self):
        if self.head != None:
            index = self.head
            nodestore = [str(index.value)]
            while index.next != None:
                index = index.next
                nodestore.append(str(index.value))
            return '[' + '->'.join(nodestore) + ']'
        return '[]'
    
        
```
