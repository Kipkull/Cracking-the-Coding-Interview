```python
class binaryTree:
    def __init__(self, content):
        self.content = content
        self.left = None
        self.right = None
        self.parent = None
        
    def __str__(self):
        if self.content == None:
            return 'None'
        return '[' + str(self.content) + \
    ':[' + 'leftChild' + ':' + str(self.left) + ',' \
    'rightChild' + ':' + str(self.right) + ']]'
        
    def hasLC(self):
        return self.left
    
    def hasRC(self):
        return self.right
    
    def isLC(self):
        return self.parent != None and self.parent.left == self
    
    def isRC(self):
        return self.parent != None and self.parent.right == self
    
    def isRoot(self):
        return self.parent == None
    
    def isLeaf(self):
        return self.left == None and self.right == None
    
    def hasC(self):
        return self.left != None or self.right != None
    
    def hasBC(self):
        return self.left != None and self.right != None
    
```

```python
class binarySearchTree:
    def __init__(self):
        self.root = None
        self.size = 0
        
    def __str__(self):
        return str(self.root)
    
    def put(self, value):
        if self.root != None:
            self._put(self.root, value)
        else:
            self.root = binaryTree(value)
            self.size += 1
    
    def _put(self, anote, value):
        if value < anote.content:
            if anote.hasLC():
                self._put(anote.left, value)
            else:
                anote.left = binaryTree(value)
                anote.left.parent = anote
        elif value > anote.content:
            if anote.hasRC():
                self._put(anote.right, value)
            else:
                anote.right = binaryTree(value)
                anote.right.parent = anote
        
    def get(self, value):
        res = self._get(self.root, value)
        if res != None:
            return res.content
        return res

    
    def _get(self, anote, value):
        if anote == None:
            return None
        if anote.content == value:
            return anote
        if anote.content < value:
            return self._get(anote.right, value)
        if anote.content > value:
            return self._get(anote.left, value)

    def __getitem__(self, value):
        return self.get(value)
        
    def __contains__(self, value):
        return self._get(self.root, value) != None
    
    def delete(self, value):
        anote = self._get(self.root, value)
        if anote != None:
            self.remove(anote)
            self.size -= 1
        else:
            return 'Error, not in tree'

        
    
    def remove(self, anote):
        if anote.isLeaf():
            if anote.isLC():
                anote.parent.left = None                
            elif anote.isRC():
                anote.parent.right = None
            else:
                anote.content = None
        elif anote.hasBC():
            succ = self.findSucc(anote)
            anote.content = succ.content
            self.sliceOut(succ)
        else:
            if anote.isLC():
                if anote.hasLC():
                    anote.parent.left = anote.left
                    anote.left.parent = anote.parent
                else:
                    anote.parent.left = anote.right
                    anote.right.parent = anote.parent
            elif anote.isRC():
                if anote.hasLC():
                    anote.parent.right = anote.left
                    anote.left.parent = anote.parent
                else:
                    anote.parent.right = anote.right
                    anote.right.parent = anote.parent
            else:
                if anote.hasLC():
                    anote = anote.left
                else:
                    anote = anote.right
                    
    def findSucc(self, anote):
        succ = self.findMin(anote.right)
        return succ
    
    def findMin(self, anote):
        curr = anote
        while curr.hasLC():
            curr = curr.left
        return curr
    
    def sliceOut(self, anote):
        if anote.isLeaf():
            if anote.isLC():
                anote.parent.left = None
            elif anote.isRC():
                anote.parent.right = None
            else:
                anote.content = None
        else:
            if anote.isRoot():
                anote = anote.right
            else:
                anote.parent.left = anote.right
                anote.right.parent = anote.parent
        
```
