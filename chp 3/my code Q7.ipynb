```python
class Node:
    def __init__(self, value, cat_or_dog=None):
        self.value = value
        self.next = None
        self.ctdg =  cat_or_dog

    def __str__(self):
        return str(self.value) + " " + self.ctdg
    
class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def addNode(self,value):
        node = Node(value)
        #if the old list is none, set new node as the first node
        if self.head == None:
            self.head = node
            self.tail = node
        else:
            self.tail.next = node
            self.tail = node

    def __str__(self):
        if self.head != None:
            index = self.head
            nodestore = [str(index.value) + " " + index.ctdg]
            while index.next != None:
                index = index.next
                nodestore.append(str(index.value) + " " + index.ctdg)
            return "LinkedList  [ " + "->".join(nodestore) + " ]"
        return "LinkedList  []"


    #remove the first node that have the same value as the given node_value
    def removeNode(self, node_value):
        current = self.head
        if current.value == node_value:
            self.head = self.head.next
        while(current.next != None):
            if current.next.value == node_value:
                current.next = current.next.next
                break
            else:
                current = current.next
```

```python
class petShulter(LinkedList):
    def __init__(self):
        self.head = None
        self.tail = None
        
    def addNode(self, anode):
        #if the old list is none, set new node as the first node
        if self.head == None:
            self.head = anode
            self.tail = anode
        else:
            self.tail.next = anode
            self.tail = anode
        
    def enqueue(self, value, cat_or_dog):
        newNode = Node(value, cat_or_dog)
        self.addNode(newNode)
        
    def dequeueAny(self):
        if self.head == None:
            return "Empty"
        rst = self.head
        self.head = self.head.next
        return rst
    
    def dequeueDog(self):
        currNode = self.head
        if currNode == None:
            return "Empty"
        if currNode.ctdg == 'dog':
            rst = currNode
            self.head = self.head.next
            return rst
        while currNode.next != None:
            if currNode.next.ctdg == 'dog':
                rst = currNode.next
                currNode.next = currNode.next.next
                return rst
            else:
                currNode = currNode.next
        return 'No more dog'
    
    def dequeueCat(self):
        currNode = self.head
        if currNode == None:
            return "Empty"
        if currNode.ctdg == 'cat':
            rst = currNode
            self.head = self.head.next
            return rst
        while currNode.next != None:
            if currNode.next.ctdg == 'cat':
                rst = currNode.next
                currNode.next = currNode.next.next
                return rst
            else:
                currNode = currNode.next
        return 'No more cat'
    
```
