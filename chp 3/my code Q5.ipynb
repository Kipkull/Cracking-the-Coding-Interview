```python
class mystack:
    def __init__(self):
        self.storage = []

    def __str__(self):
        rowstring = ''
        rowstring = rowstring + '['
        for i in range(len(self.storage)):
                rowstring = rowstring + str(self.storage[i]) + ' '
        rowstring = rowstring +']\n'
        return rowstring 
    
    def __getitem__(self, ind):
        return self.storage[ind]
        
    def push(self, value):
        self.storage.append(value)
        
    def pop(self):
        if len(self.storage) == 0:
            return "Empty"
        return self.storage.pop()
    
    def length(self):
        return len(self.storage)

    
```

```python
class myQueue:
    def __init__(self):
        self.storage = [mystack(), mystack()]
        
    def __str__(self):
        rowstring = ''
        rowstring = rowstring + '['
        for i in range(self.storage[1].length()):
            rowstring = rowstring + str(self.storage[1][-i-1]) + ' '
        for j in range(self.storage[0].length()):
            rowstring = rowstring + str(self.storage[0][j]) + ' '
        rowstring = rowstring +']\n'
        return rowstring 
        
    def push(self, value):
        self.storage[0].push(value)
        
    def pop(self):
        if self.storage[1].length() > 0:
            return self.storage[1].pop()
        while self.storage[0].length() > 0:
            self.storage[1].push(self.storage[0].pop())
        if self.storage[1].length() == 0:
            return None
        return self.storage[1].pop()
    
    def length(self):
        return self.storage[0].length() + self.storage[1].length()
```

```python
from random import randrange

aq = myQueue()

for step in range(20):
    operation = randint(10) 
    if operation < 7:
        aq.push(operation)
        print aq
    else:
        print "pop" , aq.pop(), "then", aq
```
