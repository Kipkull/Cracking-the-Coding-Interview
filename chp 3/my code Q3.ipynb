```python
class SetOfStacks:
    def __init__(self, stacksize):
        self.storage = []
        self.stacksize = stacksize
        
    def __str__(self):
        rowstring = ''
        for i in range(len(self.storage)):
            rowstring = rowstring + '['
            for vl in self.storage[i]:
                rowstring = rowstring + ' ' + str(vl)
            rowstring = rowstring +']\n'
        return rowstring     
        
    def push(self, value):
        if len(self.storage) == 0 or len(self.storage[-1]) == self.stacksize:
            self.storage.append([value])
        else:
            self.storage[-1].append(value)
            
    def popAt(self, pl=-1):
        if len(self.storage) == 0 or pl > len(self.storage):
            return "Nowhere", "Not thing"
        
        pl = pl % len(self.storage)
        
        if len(self.storage[pl]) == 0:
            self.storage.pop(pl)
            pl -= 1
            return self.popAt(pl)

        return pl, self.storage[pl].pop()
        
    def peek(self):
        if len(self.storage) == 0:
            return None
        if len(self.storage[-1]) == 0:
            self.storage.pop[-1]
            return self.peek()
        return self.storage[-1][-1]
```
